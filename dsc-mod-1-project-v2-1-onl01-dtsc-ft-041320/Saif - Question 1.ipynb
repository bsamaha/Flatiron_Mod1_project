## Q1) What is the best day/month to release movies vs popularity/domestic net


```python
# Importing the important libraries that we shall use in this investigation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
import matplotlib
sns.set_style({'grid.color': 'black', 'axes.facecolor': 'white', 'figure.facecolor': 'white',
               'axes.edgecolor': 'black','axes.labelcolor': 'white', 'xtick.color': 'black',
               'ytick.color': 'black'})
```

```python
#Getting the group_data csv file 
rel_movies = pd.read_csv("Data/group_data.csv")
display(rel_movies.info())
```

```python
#Removing Unnamed: 0 column
rel_movies.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
#Converting release_date to datetime 
rel_movies['release_date'] = pd.to_datetime(rel_movies['release_date'])
```

```python
#Adding column domestic_net to get the domestic_gross - production_budget
rel_movies['domestic_net'] = rel_movies['domestic_gross'] - rel_movies['production_budget']
rel_movies.info()
```

```python
#Visualizing release_day vs domestic net  #use show fliers = False to hide outliers

plt.figure(figsize=(13, 8))
sns.boxplot(x='release_day', y='domestic_gross', data=rel_movies)
plt.show()
```

```python
#Visualizing release_day vs popularity

plt.figure(figsize=(13, 8))
sns.boxplot(x='release_day', y='popularity', data=rel_movies)
plt.show()
```

```python
#Visualizing release_month and domestic net
plt.figure(figsize=(13, 8))
sns.boxplot(x='release_month', y='domestic_gross', data=rel_movies)
plt.show()
```

```python
#Visualizing release_month and popularity
plt.figure(figsize=(13, 8))
sns.boxplot(x='release_month', y='popularity', data=rel_movies)
plt.show()
```

# Conclusion
Friday is the best day to release a movie, in terms of both popularity and also domestic gross.                        December is the best month to release a movie,  in terms of both popularity and domestic gross.

# Insight

Try to release movies on a Friday where more people will come and watch the movie which will drive up ticket sales as compared to other days.
If the movie is originally set to release in October or November, push that date to December for maximizing gains.
